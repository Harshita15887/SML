```python
import pandas as pd
```

```python
df=pd.read_csv('/content/breast_cancer_survival.csv')
df
```

```python
df=df.fillna(0)
df
```

```python
df['ER status']=df['ER status'].replace({'Negative':0,'Positive':1})
df['PR status']=df['PR status'].replace({'Negative':0,'Positive':1})
df['HER2 status']=df['HER2 status'].replace({'Negative':0,'Positive':1})
df['Gender']=df['Gender'].replace({'MALE':0,'FEMALE':1})
df['Patient_Status']=df['Patient_Status'].replace({'Dead':0,'Alive':1})
df['Histology']=df['Histology'].replace({'Infiltrating Ductal Carcinoma':0,'Infiltrating Lobular Carcinoma':1,'Mucinous Carcinoma':2})
df['Tumour_Stage']=df['Tumour_Stage'].replace({'II':0,'I':1,'III':2})
df['Surgery_type']=df['Surgery_type'].replace({'Lumpectomy':0,'Modified Radical Mastectomy':1,'Simple Mastectomy':2,'Other':3})
```

```python
df
```

```python
df.drop('Date_of_Surgery',inplace=True,axis=1)
df.drop('Date_of_Last_Visit',inplace=True,axis=1)
```

```python
y=df['Patient_Status']
y
```

```python
x=df.drop('Patient_Status',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python

model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python

acc=accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn.svm import SVC
```

```python

svm=SVC()
svm.fit(x_train,y_train)
yp=svm.predict(x_test)
yp
```

```python
ac=accuracy_score(y_test,yp)
ac
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_p=knn.predict(x_test)
y_p
```

```python
acu=accuracy_score(y_test,y_p)
acu
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python
x_train_pca.shape
```

```python
model=LogisticRegression()
model.fit(x_train_pca,y_train)
y_pred_p=model.predict(x_test_pca)
```

```python
y_pred_p
```

```python
acp=accuracy_score(y_test,y_pred_p)
acp
```

```python
svc=SVC()
svc.fit(x_train_pca,y_train)
y_p_c=svc.predict(x_test_pca)
y_p_c
```

```python
aca=accuracy_score(y_test,y_p_c)
aca
```

```python
knn=KNeighborsClassifier()
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_pca,y_train)
y_p_a=knn.predict(x_test_pca)
y_p_a
```

```python
acca=accuracy_score(y_test,y_p_a)
acca
```
