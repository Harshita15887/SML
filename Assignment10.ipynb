```python
import pandas as pd
```

```python
df=pd.read_csv('/content/breast_cancer_survival.csv')
```

```python
df
```

```python
df=df.fillna(0)
df
```

```python
df['ER status']=df['ER status'].replace({'Negative':0,'Positive':1})
df['PR status']=df['PR status'].replace({'Negative':0,'Positive':1})
df['HER2 status']=df['HER2 status'].replace({'Negative':0,'Positive':1})
df['Gender']=df['Gender'].replace({'MALE':0,'FEMALE':1})
df['Patient_Status']=df['Patient_Status'].replace({'Dead':0,'Alive':1})
df['Histology']=df['Histology'].replace({'Infiltrating Ductal Carcinoma':0,'Infiltrating Lobular Carcinoma':1,'Mucinous Carcinoma':2})
df['Tumour_Stage']=df['Tumour_Stage'].replace({'II':0,'I':1,'III':2})
df['Surgery_type']=df['Surgery_type'].replace({'Lumpectomy':0,'Modified Radical Mastectomy':1,'Simple Mastectomy':2,'Other':3})
```

```python
y=df['Patient_Status']
y
```

```python
df.drop('Date_of_Surgery',inplace=True,axis=1)
df.drop('Date_of_Last_Visit',inplace=True,axis=1)
```

```python
x=df.drop('Patient_Status',axis=1)
x
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
test_sizes=np.arange(0.1,1.0,0.1)
# test_sizes=[i*0.1 for i in range(1,10,1)]
print(test_sizes)
for i in test_sizes:
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=i,random_state=42)
  print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
for i in range(1,21):
  knn=KNeighborsClassifier(n_neighbors=i)
knn.fit(x_train,y_train)
yp=knn.predict(x_test)
print(yp)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,yp)
accuracy
```

```python
from sklearn.svm import SVC
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# test_sizes=np.arange(0.2,0.4,0.5)
test_sizes=[20, 25, 30, 35]
acc=[]
for i in test_sizes:
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=i,random_state=42)
  print(x_train.shape,x_test.shape)
  svc=SVC()
  svc.fit(x_train,y_train)
  y_pred=svc.predict(x_test)
  y_pred
  accuracy=accuracy_score(y_test,y_pred)
  acc.append(accuracy)
  print(accuracy)
```
