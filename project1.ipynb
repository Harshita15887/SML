```python
import pandas as pd
```

```python
df=pd.read_csv('/content/student_addiction_dataset_test.csv')
df
```

```python
print(df.dtypes)
```

```python
df.head()
```

```python
df['Addiction_Class']=df['Addiction_Class'].replace({'Yes':1,'No':0})
```

```python
df['Experimentation']=df['Experimentation'].replace({'Yes':1,'No':0})
df['Academic_Performance_Decline']=df['Academic_Performance_Decline'].replace({'Yes':1,'No':0})
df['Social_Isolation']=df['Social_Isolation'].replace({'Yes':1,'No':0})
df['Financial_Issues']=df['Financial_Issues'].replace({'Yes':1,'No':0})
df['Physical_Mental_Health_Problems']=df['Physical_Mental_Health_Problems'].replace({'Yes':1,'No':0})
df['Legal_Consequences']=df['Legal_Consequences'].replace({'Yes':1,'No':0})
df['Relationship_Strain']=df['Relationship_Strain'].replace({'Yes':1,'No':0})
df['Risk_Taking_Behavior']=df['Risk_Taking_Behavior'].replace({'Yes':1,'No':0})
df['Withdrawal_Symptoms']=df['Withdrawal_Symptoms'].replace({'Yes':1,'No':0})
df['Denial_and_Resistance_to_Treatment']=df['Denial_and_Resistance_to_Treatment'].replace({'Yes':1,'No':0})
```

```python
df=df.fillna(0)
```

```python
print(df)
```

```python
y=df['Addiction_Class']
y
```

```python
df.drop('Physical_Mental_Health_Problems',inplace=True,axis=1)
df.drop('Relationship_Strain',inplace=True,axis=1)
df.drop('Risk_Taking_Behavior',inplace=True,axis=1)
```

```python
x=df.drop('Addiction_Class',axis=1)
x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.impute import SimpleImputer
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_train_resample,y_train_resample=smote.fit_resample(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train_resample, y_train_resample, test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_pred)
print(confusion)
```
