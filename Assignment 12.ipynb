```python
import pandas as pd
```

```python
df=pd.read_csv('/content/breast_cancer_survival.csv')
```

```python
df
```

```python
df['ER status']=df['ER status'].replace({'Negative':0,'Positive':1})
df['PR status']=df['PR status'].replace({'Negative':0,'Positive':1})
df['HER2 status']=df['HER2 status'].replace({'Negative':0,'Positive':1})
df['Gender']=df['Gender'].replace({'MALE':0,'FEMALE':1})
df['Patient_Status']=df['Patient_Status'].replace({'Dead':0,'Alive':1})
df['Histology']=df['Histology'].replace({'Infiltrating Ductal Carcinoma':0,'Infiltrating Lobular Carcinoma':1,'Mucinous Carcinoma':2})
df['Tumour_Stage']=df['Tumour_Stage'].replace({'II':0,'I':1,'III':2})
df['Surgery_type']=df['Surgery_type'].replace({'Lumpectomy':0,'Modified Radical Mastectomy':1,'Simple Mastectomy':2,'Other':3})
```

```python
df
```

```python
df.drop(['Date_of_Last_Visit','Date_of_Surgery'],axis=1,inplace=True)
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
y=df['Patient_Status']
y
```

```python
x=df.drop(['Patient_Status'],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
yp=model.predict(x_test)
yp
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
log_acc=accuracy_score(y_test,yp)
log_acc
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_p=model.predict(x_test)
y_p
```

```python
svm_accu=accuracy_score(y_test,y_p)
svm_accu
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier(n_neighbors=3)
kn.fit(x_train,y_train)
y_pred=kn.predict(x_test)
y_pred
```

```python
knn_acc=accuracy_score(y_test,y_pred)
knn_acc
```

```python
from sklearn.utils import resample
```

```python
knn_res=KNeighborsClassifier(n_neighbors=3)
lr_res=LogisticRegression(max_iter=1000)
svm_res=SVC(kernel='linear')
```

```python
knn_acc=[]
l_acc=[]
s_acc=[]
```

```python
y.shape
```

```python
x.shape
```

```python
for i in range(20):
  X,Y=resample(x,y)
  x_tr,x_te,y_tr,y_te=train_test_split(X,Y,test_size=0.2,random_state=42)
  knn_res.fit(x_tr,y_tr)
  y_pr=knn_res.predict(x_te)
  knn_acc.append(accuracy_score(y_te,y_pr))
  lr_res.fit(x_tr,y_tr)
  y_predi=lr_res.predict(x_te)
  l_acc.append(accuracy_score(y_te,y_predi))
  svm_res.fit(x_tr,y_tr)
  y_predict=svm_res.predict(x_te)
  s_acc.append(accuracy_score(y_te,y_predict))
```
